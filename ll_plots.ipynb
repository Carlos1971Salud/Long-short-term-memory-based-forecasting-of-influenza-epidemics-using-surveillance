```python
import matplotlib.pyplot as plt

# 
x_pos = list(range(5))  # : [0, 1, 2, 3, 4]
x_labels = [1, 4, 13, 26, 52]  # 

#  log-likelihood 
vanilla_lstm   = [-7.5e6, -6.1e6, -5.1e6, -4.8e6, -4.9e6]
auxiliary_lstm = [-8.0e6, -5.1e6, -4.9e6, -4.1e6, -4.7e6]
vector_lstm    = [-4257786, -3980212, -3679987, -4686962, -4009568]
auxvec_lstm    = [-3570094, -3665872, -3472671, -3508939, -4863001]

# 
fig, ax = plt.subplots(figsize=(6, 6))

# x_posx_labels
ax.plot(x_pos, vanilla_lstm,   linestyle='--', marker='s', markeredgecolor='gray',color='blue', label='Vanilla LSTM')
ax.plot(x_pos, auxiliary_lstm, linestyle='--', marker='s', markeredgecolor='gray',color='darkorange', label='Auxiliary LSTM')
ax.plot(x_pos, vector_lstm,    linestyle='--', marker='s', markeredgecolor='gray',color='green', label='Vector LSTM')
ax.plot(x_pos, auxvec_lstm,    linestyle='--', marker='s', markeredgecolor='gray',color='red', label='Aux-vec LSTM')

# 
ax.set_xlabel("Length of input sequence", fontsize=16)
ax.set_ylabel("Log-likelihood", fontsize=16)

# x
ax.set_xticks(x_pos)
ax.set_xticklabels(x_labels)

# y
ax.ticklabel_format(style='sci', axis='y', scilimits=(6, 6))

ax.tick_params(axis='x', labelsize=12)
ax.tick_params(axis='y', labelsize=12)

# 
ax.grid(False)
ax.legend(loc='lower right', fontsize=14)
ax.yaxis.get_offset_text().set_fontsize(13)

# 
plt.tight_layout()

# 
plt.grid()
plt.savefig("ll_plot.png", dpi=300, bbox_inches='tight')

# 
# 
plt.show()
```
